# Co-prime Numbers: 

* Two positive integers $(p,q)$ are said to be co-prime if their greatest common divisor (GCD) is 1. OR
* Two positive integers $(p,q)$ are **co-prime** or **relatively prime** if they do not have a common divisor. Hence $GCD(p,q) = 1$.

Using `Numpy` to find co-prime numbers `np.gcd()`:  

```python
import numpy as np 
import math, cmath 
```

```python
def iscoprime(p=1,q=1):
    common_factor = np.gcd(p,q)
    if np.gcd(p,q)==1:
        return (True, common_factor)
    else:
        return (False, common_factor)
    
p = int(input("First Positive Integer: "))
q = int(input("Second Positive Integer: "))

print("Are ",p," and ",q," co-prime? : ",iscoprime(p,q)) 
```
