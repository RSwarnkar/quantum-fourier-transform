# QFT Simulation using Qiskit

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, Aer, execute 
from qiskit.visualization import plot_bloch_multivector, plot_bloch_vector
from qiskit.circuit.library import QFT
```

```python
num_qubits = 5
qc = QFT(num_qubits=num_qubits, do_swaps=True, inverse=False, insert_barriers=True, name='qft_5')

qc.decompose().draw("mpl", scale=1, fold=False, reverse_bits=True)
```

```python
job = execute(qc, backend=Aer.get_backend("statevector_simulator"), shots=1000)
counts = job.result().get_counts()
print(counts)
```

```python
qc.decompose().depth()
```

```python
qc.width()
```

### Inverse QFT: 

```python
num_qubits = 4
iqc = QFT(num_qubits=num_qubits, do_swaps=True, inverse=True, insert_barriers=True, name='iqft')

iqc.decompose().draw("mpl", scale=1, fold=False, reverse_bits=True)
```

```python
job = execute(iqc, backend=Aer.get_backend("statevector_simulator"), shots=1000)
counts = job.result().get_counts()
print(counts)
```

```python
qc.decompose().depth()
```
