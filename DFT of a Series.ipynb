# DFT Fourier Transform:



```python
import numpy as np 
import math, cmath
import matplotlib.pyplot as plt
```

```python
x=[]
N=100
for i in range(N):
    if i%6==0:
        x.append(1)
    else:
        x.append(0)

x = np.fft.fft(np.array(x))

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(8, 3))
axes[0].plot(x.real)
axes[1].plot(x.imag)
axes[2].plot(x)
fig.tight_layout()
```
